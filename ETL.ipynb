```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import missingno as msno
import seaborn as sns
```

```python
# acceso a internet por tecnología y provincia
internet_tecnología_provincia= pd.read_csv("internet/Acceso_Internet_fijo_por_tecnología_y_provincia.csv")
```

```python
#creamos una columna item, que se usará para unir dataframe
internet_tecnología_provincia['Item'] = internet_tecnología_provincia.Año.astype(str) + internet_tecnología_provincia.Provincia.astype(str)
internet_tecnología_provincia.head()
```

```python
#penetración de internet fijo, accesos cada 100 hogares
penetracion_internet=pd.read_csv("internet/Penetración_de _Internet fijo _(accesos _cada_100 hogares).csv")
```

```python
penetracion_internet['Item'] = penetracion_internet.Año.astype(str) + penetracion_internet.Provincia.astype(str)
penetracion_internet.head()
```

```python
# Selecciono solo las columnas que quiero para le merge
tecnologia_penetracion_merged = pd.merge(internet_tecnología_provincia, penetracion_internet, how='left', on='Item')
tecnologia_penetracion_merged.head()
```

```python
# Exporto a csv el dataframe
tecnologia_penetracion_merged.to_csv('dataset_etl/tecnologia_penetracion_merged.csv', index=False)
```
